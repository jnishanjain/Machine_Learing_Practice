```python
import numpy as np
```

```python
np.random.seed(0)
X = np.random.randint(1,10,(100000,10))
```

```python
X.shape
```

```python
#creating Y
weights = np.array([5,6,3,1,2,5,6,7,8,10])
bias = 4
y = np.dot(X,weights)
```

```python
y.shape
```

```python
class Linearregression():
  def __init__(self,max_iteration = 10000,learning_rate = 0.001):
    self.max_iter = max_iteration
    self.lr = learning_rate
    self.loss_history = []
    self.weights = None
    self.bias = None

  def fit(self,X,y):

    n_samples , n_features = X.shape
    self.weights = np.zeros(n_features)
    self.bias = 0

    for epochs in range(self.max_iter):

      y_pred = self.predict(X)
      error = y_pred - y
      dw = np.dot(X.T,error)/n_samples
      db = np.sum(error)/n_samples
      self.weights = self.weights - self.lr*dw
      self.bias = self.bias - self.lr*db
      loss = np.mean(error**2)
      self.loss_history.append(loss)
      if epochs % 1000 == 0 :
        print(f"Epoch : {epochs} , Loss : {loss}")
  def predict(self,X):
    return np.dot(X,self.weights) + self.bias
```

```python
model = Linearregression()
model.fit(X,y)
```

```python
print(f"Orignal weights and bias ")
print(f"weights :{weights}")
print(f"bias :{bias}")
print(f"Predicted weights and bias ")
print(f"Predicted weights :{model.weights}")
print(f"Predicted bias :{model.bias}")
```

```python
import matplotlib.pyplot as plt
plt.plot(model.loss_history)
plt.show()
```
