```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
 #Setting up our random seed so out results are reproducable
np.random.seed(59)
X =  np.random.randint(1,10,(10000,2))
X.shape
```

```python
weight_1 = 10
weight_2 = -5
bias = 10
y = X[:,0]*weight_1 + X[:,1]*weight_2 + bias
y.shape
```

```python
class LinearRegression :
  def __init__(self,learning_rate=0.01,max_iterations = 1000):
    self.lr = learning_rate
    self.max_iter = max_iterations
    self.weights = None
    self.bias = None
    self.loss_history=[]

  def fit(self,X,y):
    n_samples,n_features = X.shape
    self.weights = np.zeros(n_features)
    self.bias = 0
    for epoch in range(self.max_iter):
      y_pred = self.predict(X)
      error = y_pred - y
      dw = 2/n_samples * np.dot(X.T,error)
      db = 2/n_samples * np.sum(error)
      self.weights = self.weights - self.lr*dw
      self.bias = self.bias - self.lr*db
      loss = np.mean(error**2)
      self.loss_history.append(loss)
      if epoch%1000 == 0 :
        print(f"Epoch {epoch} : Loss {loss}")
  def predict(self,X):
     return np.dot(X,self.weights) + self.bias
```

```python
model = LinearRegression(learning_rate=0.001,max_iterations=10000)
model.fit(X,y)
```

```python
print("Orignal Weights")
print(f"Weight 1 : {weight_1} , Weight 2 : {weight_2}")
print("Predicted Weights")
print(f"Weight 1 : {model.weights[0]}, Weight 2 : {model.weights[1]}")
print("Orignal Bias")
print(f"Bias : {bias}")
print("Predicted Bias")
print(model.bias)
```
